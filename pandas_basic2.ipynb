```python
# 분석하기 좋은 데이터

# 분석하기 좋은 데이터는 데이터 집합을 분석하기 좋은 상태로 만들어놓은 것을 말한다.
# 데이터 분석에서는 데이터 정리가 아주 중요하다. 실제로 데이터 분석 작업의 70~80% 를 
# 차지하는 부분이 데이터 정리 작업이다 (전처리 : preprocessing)

# 깔끔한 데이터 Tidy Data
# - 데이터 분석 복적에 맞게 데이터를 수집하여 표(데이터프레임)로 만들어야한다
# - 같은 행으로 구성한다
# - 변수는 열로 구성한다 
```

```python
# 데이터 연결 

import pandas as pd

df1 = pd.read_csv('concat_1.csv')
df2 = pd.read_csv('concat_2.csv')
df3 = pd.read_csv('concat_3.csv')

# 판다스로 엑셀이나 csv 를 가져오면 데이터프레임으로 만든다 

type(df1)
```

```python
# 가져온 세개의 데이터프레임을 하나로 연결한다

row_concat = pd.concat([df1, df2, df3])
row_concat
```

```python
new_row_series = pd.Series(['n1', 'n2', 'n3', 'n4'])

# 시리즈와 데이터프레임을 연결 

pd.concat([df1, new_row_series])

# 데이터를 전처리하다보면 의도하지 않게 결측되는 데이터가 빈번하게 발생한다 
```

```python
i_row_concat = pd.concat([df1, df2, df3], ignore_index=True)
i_row_concat

# 기존 인덱스 체계는 삭제되고 새로운 인덱스가 전체에 부여된다 
```

```python
# 열 방향으로 데이터 연결하기
col_concat = pd.concat([df1, df2, df3] , axis = 1)    # axis = 1 : 열방향, 생략하면 행방향 
col_concat
```

```python
col_concat['A']
```

```python
col_concat['F'] = ['1', '2', '3', '4']
col_concat    # 없는 열이라면 만들어서 집어넣는다 
```

# 결측치 처리 

분석할 데이터에 결측치(null)는 대부분 존재한다. 분석이나 머신러닝 전에 결측치를 처리하는 것은 매우 중요함.

결측치는 NaN, NAN, nan 으로 표시할 수 있다. 

```python
# 결측치 사용을 위해 numpy 임포트

from numpy import NaN, NAN, nan

# 결측치는 0, '' 즉 영과 빈칸과는 다른 것이다. 
# -> 데이터 자체가 없다는 의미이다.
# 따라서 같다(=)라는 개념도 없다 

print(NaN == True)
print(NaN == False)
print(NaN == '')
print(NaN == 0)
print(NaN == NaN)
print(NaN == NAN)
print(NaN == nan)
print(pd.isnull(NaN))   # 이렇게는 찾을 수 있다 
```

# 결측치 처리하기

1. 결측치가 존재하는 행 전부를 삭제한다. (데이터의 개수가 아주 많을 때) 
2. 결측치를 그 열의 평균값으로 대체 (데이터가 숫자 일 때)
3. 결측치를 그 열의 최빈값으로 대체 (데이터가 문자/카테고리 일 때)
4. 결측치를 0으로 대체한다 (외곡이 심함)
