# 인공지능(artificial intelligence)  
>> 사람처럼 학습하고 추론할 수 있는 지능을 가진 컴퓨터 시스템 또는 이러한 기술.

### 머신러닝(machine learning)  
>> 직접 규칙을 프로그래밍 하지 않아도 데이터에서 규칙을 자동으로 학습하는 알고리즘 연구 분야.
	현재도 개발중인 알고리즘이 많고, 이미 개발된 알고리즘을 가져다 사용하게 된다.
	대표적인 머신러닝 라이브러리는 사이킷런(scikit-learn)이며,
	모든 머신러닝 알고리즘이 포함되어 있지 않고 유익하다고 증명되어 많은 사람들이
	사용하게 되었을 때 비로소 사이킷런 라이브러리 개발자들이 해당 알고리즘을
	라이브러리에 추가하게 된다. 그러므로 머신러닝 라이브러리에 포함된 알고리즘들은
	안정적이며 성능이 검증되어 있다.
	개발자는 직접 알고리즘을 구현할 필요 없이 검증된 알고리즘이 들어있는 머신러닝
	라이브러리를 통해 사용하기만 하면 된다.
	머신러닝 기술들은 대부분 폐쇄적인 코드와 라이브러리로 통용되었고 전문교육 이수
	또는 고가의 금액으로 구매해야 사용 가능하였지만, 사이킷런이라는 오픈소스
	라이브러리의 발전덕에 머신러닝 분야가 인기를 얻고 있다.

### 딥러닝(deep learning)  
>> 머신러닝 알고리즘 중 인공 신경망(articicial neural network)을 기반으로 한 방법들.
	구글의 텐서플로(TensorFlow), 페이스북의 파이토치(PyTorch)가 대표적인 딥러닝 오픈소스
	라이브러리이다.
	*인공신경망 : 인간의 뇌가 학습하는 알고리즘을 분석하여 만들어낸 통계학적 학습 알고리즘

### 아나콘다 설치(Anaconda)  
>> 파이썬은 기본적으로 패키지 관리 시스템인 pip만 포함하고 있어서 필요한 패키지를 수동으로 추가해야 한다.

   하나의 방에 여러 패키지가 섞이지 않고 패키지의 성격별로 관리하기 때문에 가상 환경 관리자이기도 하다.
